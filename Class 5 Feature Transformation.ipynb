# Advantages of Feature Scaling

```python
# Make the training faster
# Prevents from getting stuck in local optima
# Provide a better error surface shape

# Probability based algorithm and tree based algorithm are not affected by scaling of input data( Naive Bayes, Desicion Tree, Random Forest)
```

# Scaling matters:

```python
# K - Means
# K - Nearest - Neighbours
# Principal Component Analysis(PCA)
# Gradient Descent
```

# Techniques to perform feature scaling:

```python
# 1) Normalization  (Widely Used)

# 2) Standardization  (Widely Used)

# 3) Robust Scaler (Contain out layer)

# 4) Max Absolute Scaler 
# In Max Absolute Scaler - Find the large value and divide each value of the column like (40/60) or (50/60) or (60/60)
# Scaling value must be within the limit( 0 - 1)
```
